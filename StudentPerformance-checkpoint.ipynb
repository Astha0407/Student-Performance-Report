```python
df['total marks']=df['math score']+df['reading score']+df['writing score']
```

```python
df['percentage']=df['total marks']/300*100
```

```python
def determine_grade(scores):
    if scores>=85 and scores<=100:
        return 'Grade A'
    elif scores>=70 and scores<85:
        return 'Grade B'
    elif scores>=55 and scores<69:
        return 'Grade C'
    elif scores>=35 and scores<54:
        return 'Grade D'
    elif scores>=0 and scores<35:
        return 'Grade E'
    return scores
pass
df['scores']=df['percentage']
df['grades']=df['determine_grade']
```

```python
df.info()
```

```python
df['grades'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.subplots_adjust(left=0.125,bottom=0.1,right=0.9,wspace=0.5,hspace=0.2)
plt.subplot(251)
plt.title('percentace and gender')
sns.countplot(hue="gender",x="grades",data=df)
plt.subplot(252)
plt.title('grades and lunch')
sns.countplot(hue="lunch",x="grades",data=df)
plt.subplot(253)
plt.title('grades and test preparation course')
sns.countplot(hue="test preparation course",x="grades",data=df)
plt.show()
```
